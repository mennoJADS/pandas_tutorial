Rename all colums such that:
1. all names are in lowercase
2. all spaces are replaces with underscores
3. any whitespace in the front or back is removed so: 

Example:
" Column Name " --> "column_name"

```python
#Option one:
chicago.columns = chicago.columns.str.strip().str.lower().str.replace(' ', '_')
chicago.head(1)

#option two:
for column in chicago.columns.values:
    chicago = chicago.rename(index=str, columns={column:str(column.lower().replace(" ", "_").strip())})
chicago.head(1)
```

Operators: instead of 'and', 'or' pandas uses & for and,  | for or.
List the salaries of seargants of the police department, order them from high to low income.

```python
chicagofiltered = chicago[(chicago['department'] == 'POLICE') & (chicago['job_titles'] == 'SERGEANT')]\
    ['annual_salary'].sort_values()
chicagofiltered.head(5)
```

Sort departments based on descending order.

```python
chicagosortdept = chicago['department'].sort_values(ascending=False)
chicagosortdept.head()
```

Transfer the columns "Full or Part-Time" and "Salary or Hourly" to binary 0 and 1 values.

```python
#lambda x: True if x % 2 == 0 else False
chicago['salary_or_hourly'] = chicago.apply(lambda x: 0 if str(x['salary_or_hourly']) == 'Hourly'\
                                            else 0, axis=1)
chicago['full_or_part-time'] = chicago.apply(lambda x: 0 if str(x['full_or_part-time']) == 'F'\
                                            else 0, axis=1)
chicago.head()
```

Separate the names on the comma ','

```python
chicago[['fistname','lastname']] = chicago['name'].str.split(',',expand=True)
chicago.head()
```
